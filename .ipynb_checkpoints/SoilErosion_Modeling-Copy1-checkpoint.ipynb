```python
X_train = []
y_train = []
num_samples = len(train_img)
image_shape = 64

for idx in range(0, num_samples, image_shape):
    x_ = train_img[idx:idx + image_shape, idx:idx + image_shape,...] 
    y_ = test_img[idx:idx + image_shape, idx:idx + image_shape,...]
    X_train.append(x_)
    y_train.append(y_)
    if len(y_) < image_shape: 
        break
```

```python
inputs = layers.Input(shape=(64, 64, 3))
net = layers.Conv2D(32, kernel_size=(3, 3), activation="relu", padding='same')(inputs)
net = layers.MaxPooling2D(pool_size=(2, 2))(net)
net = layers.Dropout(0.2)(net)
# net = layers.Conv2D(64, kernel_size=(3, 3), activation="relu", padding='same')(net)
net = layers.MaxPooling2D(pool_size=(2, 2), activity_regularizer=tf.keras.regularizers.L2(0.01))(net)
net = layers.Dropout(0.2)(net)
# net = layers.Dense(64)(net)
net = layers.UpSampling2D((4, 4))(net)
outputs = layers.Conv2D(1, (3, 3), activation='sigmoid', padding='same')(net)

model = Model(inputs, outputs)
model.summary()
```

```python
from tensorflow import keras
from sklearn.utils import shuffle

class Generator(keras.utils.Sequence):    
    def __init__(self, batch_size, img_size, X, y):
        self.batch_size = batch_size
        self.img_size = img_size
        self.X = X
        self.y = y

    def __len__(self):
        return len(self.img_size) // self.batch_size

    def __getitem__(self, idx):
        i = idx * self.batch_size
        x_ = self.X[i : i + self.batch_size, i : i + self.batch_size,...] 
        y_ = self.y[i : i + self.batch_size, i : i + self.batch_size,...]
        if len(y_) < batch_size: 
            x_train, y_train = shuffle(x_train, y_train) 
            i = idx * self.batch_size
            x_ = self.X[i : i + self.batch_size, i : i + self.batch_size,...] 
            y_ = self.y[i : i + self.batch_size, i : i + self.batch_size,...]
            
        return x_, y_
    
    def on_epoch_end(self):
        self.X, self.y = self.X, self.y
```
